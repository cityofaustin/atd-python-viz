```python
import plotly.express as px
import numpy as np
import pandas as pd
import chart_studio

df = pd.read_excel("ATD FY20_21 Det Exp v Budget.xlsx")

#df = df[df['Unit Name'].isin(['Office of the Director','Corridor Program Office'])]

df = df[df['Budget 2021'] > 0]

df = df[df['Unit Name'] != 'Transfers']


fig = px.treemap(df, 
                 path=
                     [
                      px.Constant("Austin Transportation Department 2021 Budget"), 
                      'Unit Name',
                      'Object Code Category',
                      'Obj Code'
                     ], 
                 values='Budget 2021',
                 maxdepth = 2)
#fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.update_traces(root_color="lightgrey")
fig.show()
```
